```python
#What are the mean, maximum and minimum salaries?
import warnings
warnings.filterwarnings('ignore')
dataframe['Salary']=dataframe['Annual Salary'].str.replace('$','').str.replace(',','').astype(float)
```

```python
dataframe.head()
```

```python
print('maximum Salary are:' ,dataframe['Salary'].max())
print('minimum Salary are:',dataframe['Salary'].min())
print('the average Salary are:',dataframe['Salary'].mean())
```

```python
# Find the employee who has the maximum salary?
dataframe[dataframe['Salary']==dataframe['Salary'].max()]['Name']
```

```python
# Find the employee who has the minimum salary?
dataframe[dataframe['Salary']==dataframe['Salary'].min()]['Name']

```

```python
#What are the max, min and mean hourly rate?
dataframe['H-rate']=dataframe['Hourly Rate'].str.replace('$','').astype(float)
```

```python
print('maximum H-rate are:' ,dataframe['H-rate'].max())
print('minimum H-rate are:',dataframe['H-rate'].min())
print('the average H-rate are:',dataframe['H-rate'].mean())
```
